```python
import sys

class allroutes:
    route = []


#create new instance of allroutes and pass on.

routecount = 0

global nodeexpansioncount
nodeexpansioncount = 0


Place_names = ["Arad" ,
               "Bucharest",
               "craiova",
               "Dobreta",
               "Eforie",
               "Fagaras",
               "Giurgiu",
               "Hirsova",
               "Iasi",
               "Lugoj",
               "Mehadia",
               "Neamt",
               "Oradea",
               "Pitesti",
               "Rimnicu Vilcea",
               "Sibiu",
               "Timisoara",
               "Urziceni",
               "vaslui",
               "Zerind"
               ]

straightline_distance = [366,0,160,242,161,178,77,151,226,244,241,234,380,98,193,253,329,80,199,374]

global straightline_distance
global graph

graph = [
            [[19,75],[15,140],[16,118]],
            [[17,98],[6,90],[5,211],[13,101]],
            [[13,101],[14,146],[3,120]],
            [[2,120],[10,75]],
            [[7,86]],
            [[1,211],[15,99]],
            [[1,90]],
            [[4,86],[17,98]],
            [[11,87],[18,92]],
            [[10,70],[16,111]],
            [[3,75],[9,70]],
            [[8,87]],
            [[15,151],[19,71]],
            [[1,101],[2,138],[14,97]],
            [[13,97],[2,146],[15,80]],
            [[14,80],[0,140],[12,151],[5,99]],
            [[0,118],[9,111]],
            [[1,85],[7,98],[18,142]],
            [[17,142],[8,92]],
            [[0,75],[12,71]]
        ]


global options


options = [[19,374]]


#print graph[0][0][1] #cost for path from place_names[0] to place_names[graph[0][0][0]] is graph[0][0][1]

def heuristic(node):
    h = straightline_distance[node]
    return h

def g(fromnode,node):
    #get cost from graph for going from fromnode to node. (straight line distances)
    cost = 0
    for i in graph[fromnode]:
        if(i[0]==node):
            cost = i[1]
    return cost

def get_possible_states(fromnode):
    possible = []
    for i in graph[fromnode]:
        possible.append(i[0])
    return possible

def getf(fromnode,node):
    global nodeexpansioncount
    nodeexpansioncount += 1
    f = g(fromnode,node) + heuristic(node)
    return f

def notinroute(node,route):
    flag = True
    for i in route:
        if(node==i):
            flag = False        
    return flag
    

def best(node,route):#checks the options available and returns the best node to expand on.
    m = 1000
    b = 0
    for i in range(0,len(options)):
        if(m>=options[i][1] and node!=options[i][0] and notinroute(options[i][0],route)):
            m = options[i][1]
            b = options[i][0]
    return b;
        
def connectedto(fromnode,bestchoice):#checks if the bestchoice from options is connected to the current expanded node or not.
    flag = False
    for i in graph[fromnode]:
        if(i[0]==bestchoice):
            flag = True
    return flag

def removeparent(node):# the parent node which is expanded now, has to be removed from the options.
    #remove from options, element [x,y] where x = node.
    
    count = 0
    temp = 0
    for i in range(0,len(options)):
        count = count +1
        if(node==options[i][0]):
            temp = options[i][1]
    temp2 = [node,temp]  
    options.remove(temp2)
    return True

def addtooptions(node):# the funcition for adding [node,f(x)] values to the options.
    flag = True
    for i in range(0,len(options)):
        if(node[0]==options[i][0]):
            options[i][1] = node[1]
            flag = False
    if(flag==True):
        options.append(node)
    
def calculatecost(arr):
    totalcost = 0
    left = arr[0]
    for i in range(1,len(arr)):
        for j in graph[left]:
            if(arr[i]==j[0]):
                totalcost = totalcost + j[1]
        left = arr[i]
    return totalcost;

def pathwithnames(arr):
    sentence = ""
    for i in arr:
        sentence = sentence + Place_names[i] + " - "
    return sentence    
        
    
    
def Astar(fromnode, routeinstance, goal):
    removeparent(fromnode)
    if(fromnode==goal):
        return 0;
    
    possible_states = get_possible_states(fromnode)
    for i in possible_states:
        temp = getf(fromnode,i)
        temp2 = [i,temp]
        addtooptions(temp2)
    print "options: ",options
    bestchoice = best(fromnode,routeinstance.route);   
    print "bestchoice" , bestchoice
    #if bestchoice is not with a child of this node, then remove this node from route.
    if (connectedto(fromnode,bestchoice)==False):
        print fromnode,"not connected to ",bestchoice,"so removing",fromnode, "from route"
        routeinstance.route.remove(fromnode)
            
    routeinstance.route.append(bestchoice)
    Astar(bestchoice,routeinstance, goal)
    
    

path = allroutes()

path.route = [19]

print "id) PlaceName"," :  Straightline distances to Bucharest"
for i in range(0,20):
    print i, ") ", Place_names[i],  " : " , straightline_distance[i]
 
Astar(19,path,1)

print "Shortest cost path is: ",
print pathwithnames(path.route)
print "Total cost is: ",calculatecost(path.route)
print "total no of nodes expanded in algorithm:" , nodeexpansioncount
   
```
